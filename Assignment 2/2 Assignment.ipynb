```python
class Node(object):
    def __init__(self, data, n=None):
        self.data = data
        self.next_node = n

    def __str__(self):
        return str(self.data)
```

```python
class LinkedList(object):
    def __init__(self, r=None):
        self.head = r

    def __str__(self):
        current = self.head
        output = ""
        while current:
            output += str(current.data) + " "
            current = current.next_node
        return output + "None"

    def add(self, d):
        new_node = Node(d, self.head)
        self.head = new_node

    def append(self, d):
        new_node = Node(d)

        if self.head is None:
            self.head = new_node
            return

        current = self.head

        while current.next_node is not None:
            current = current.next_node

        current.next_node = new_node

    def pop(self):
        if self.head is None:
            return

        if self.head.next_node is None:
            self.head = None
            return
        current = self.head

        while current.next_node.next_node is not None:
            current = current.next_node

        current.next_node = None

    def find(self, t):

        current = self.head

        while current is not None:
            if current.data == t:
                return True

            current = current.next_node

        return None

    def insert(self, d, pos):
        count = 0
        new_node = Node(d)

        if pos == 0:
            self.add(d)
            return 

        current = self.head
        while current is not None and count < pos-1:
            current = current.next_node
            count += 1

        if current is None:
            return
        new_node.next_node = current.next_node

        current.next_node = new_node

    def __iter__(self):
        current = self.head
        while current:
            yield current.data
            current = current.next_node

    def remove_by_value(self, data):

        current = self.head
        if self.head.data == data:
            self.head = self.head.next_node
            return

        while current.next_node is not None and current.next_node.data != data:
            current = current.next_node

        if current.next_node is not None:
            current.next_node = current.next_node.next_node

    def merge_list(self, new_linked_list):

        current = self.head
        while current.next_node is not None:
            current = current.next_node

        for i in new_linked_list:
            self.append(i)

    def reverse(self):
        a = []
        for i in self:
            a.append(i)
        a.reverse()

        b = LinkedList()

        for j in a:
            b.append(j)

        return b

    def sort(self):
        a = []
        for i in self:
            a.append(i)
        a.sort()
        b = LinkedList()
        for j in a:
            b.append(j)

        return b
```

```python
n1 = Node(1)
a = LinkedList(n1)
a.add(8)
a.add(7)
a.add(6)
print(a)
```
