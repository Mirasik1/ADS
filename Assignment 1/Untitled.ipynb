```python
#Task 1
def print_numbers(n):
    if n == 0:
        return
    print_numbers(n-1)
    print(n)
print_numbers(5)
```

```python
#Task 2
def print_numbers(n):
    print(n)
    if n == 1:
        return
    print_numbers(n-1)
print_numbers(5)
```

```python
#Task 3
def sub_factorial(n):
    if n==1:
        return 1
    
    return n+sub_factorial(n-1)
print(sub_factorial(5))
```

```python
#Task 4
def factorial(n):
    if n==1:
        return 1
    return n*factorial(n-1)
print(factorial(5))
```

```python
#Task 5
def power(a,b):
    if b==1:
        return a
    return a*power(a,b-1)
print(power(5,4))
```

```python
#Task 6
def sum_of_digits(n):
    if n== 0:
        return 0
    return n%10 + sum_of_digits(int(n/10))
print(sum_of_digits(572))
```

```python
#Task 7
def count_of_digits(n):
    if n == 0:
        return 0
    return 1+count_of_digits(int(n/10))
print(count_of_digits(7655))
```

```python
#Task 8
def reverse_digits(n):
    if n == 0:
        return ''
    return str(n%10) + str(reverse_digits(int(n/10)))
print(reverse_digits(572))
```

```python
#Task 9
def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fib(n-1) + fib(n-2)
print(fib(6))
```

```python
#Task 10
def polin(t):
    if len(t) <1:
        return "Polindrom"
    if t[0] != t[-1]:
        return "Not Polindrom"
    return polin(t[1:-1])
polin("level")
```

```python
#Task 11
a= [4,9,1,7,3]
def sum_arr(arr):
    if len(arr) == 0:
        return 0
    return sum_arr(arr[1:]) + arr[0]
print(sum_arr(a))
```

```python
#Task 12
a= [4,9,1,7,3]
def maximum(a):
    
    if len(a) == 1:
        return a[0]
    if a[-1] > maximum(a[:-1]):
        return a[-1]
    else:
        return maximum(a[:-1])
print(maximum(a))
```

```python
#Task 13
a= [4,9,1,7,3,3,3]
def find_in_arr(a,t):
    if len(a) == 0:
        return 0
    if a[0] == t:
        return find_in_arr(a[1:],t)+1
    return find_in_arr(a[1:],t)
print(find_in_arr(a,3))
```

```python
#Task 14
a= [4,7,1,9,3]
def linear_search(a, b):
    if len(a) == 0:
        return "Not Found"
    if a[0] == b:
        return "Found"
    return linear_search(a[1:],b)
print(linear_search(a,1))
```

```python
#Task 15
a= [1,2,3,4,5,6,7]
def is_sorted(a):
    print(a)
    if len(a) == 1:
        return "Sorted"
    if a[0] >= a[1]:
        return "Not Sorted"
    return is_sorted(a[1:])
print(is_sorted(a))
```
